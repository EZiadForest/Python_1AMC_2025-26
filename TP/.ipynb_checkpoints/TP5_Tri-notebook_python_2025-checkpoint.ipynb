## Tri rapide (Quicksort)

Référence pour la définition: http://fr.wikipedia.org/wiki/Quicksort

````python
def trirapide(L):
    """trirapide(L): tri rapide (quicksort) de la liste L"""
    def trirap(L, g, d):
        if g >= d:
            return
        pivot = L[(g + d) // 2]
        i, j = g, d
        while i <= j:
            while L[i] < pivot:
                i += 1
            while L[j] > pivot:
                j -= 1
            if i <= j:
                L[i], L[j] = L[j], L[i]
                i += 1
                j -= 1
        trirap(L, g, j)
        trirap(L, i, d)

    copied_list = list(L)
    trirap(copied_list, 0, len(copied_list) - 1)
    return copied_list
````

````python
# Exemple d'utilisation
tableau = [64, 34, 25, 12, 22, 11, 90]
tableau_trié = trirapide(tableau)
print("Tableau trié :", tableau_trié)
````

Le pivot est choisi comme l'élément médian de la sous-liste (L[(g + d) // 2]).   
C'est une méthode simple et couramment utilisée, mais il existe d'autres stratégies de choix du pivot qui peuvent améliorer les performances dans certains cas (par exemple, choisir le premier, le dernier ou un élément aléatoire).  
```algo
tri_rapide_alea(Tableau T)
    fonction tri_rap(Tableau T, entier g, entier d)
        si g < d
            pivot_index := aléatoire(g, d)
            pivot := T[pivot_index]
            échanger(T[pivot_index], T[d])
            i := g
            j := d - 1
            tant que vrai
                tant que i <= j et T[i] <= pivot
                    i := i + 1
                tant que i <= j et T[j] >= pivot
                    j := j - 1
                si i >= j
                    interrompre
                échanger(T[i], T[j])
            échanger(T[d], T[i])
            tri_rap(T, g, i - 1)
            tri_rap(T, i + 1, d)

    g := 0
    d := taille de T - 1
    tri_rap(T, g, d)
    retourner T
```
